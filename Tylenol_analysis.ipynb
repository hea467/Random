```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open("oxycodone.txt") as f:
    lines = f.readlines()
    resx=[]
    resy=[]
    countres=[]
    app=False
    count=0
    for line in lines:
        if line.startswith("[PDA"):
            app=False
            countres.append(count)
            count=0
        if line.strip()=="":
            app=False
            continue
        if app:
            tmp=line.split("	")
            resx.append(float(tmp[0]))
            resy.append(float(tmp[1]))
            count +=1
        if line.startswith("R.Time"):
            app=True
```

```python
plt.plot(resx[0:9476], resy[0:9476], color="blue") #205
plt.plot(resx[9476:(9476*2)], resy[9476:(9476*2)], color="green") #250 
plt.plot(resx[9476*2:(9476)*3], resy[9476*2:(9476*3)], color="orange") #290
plt.xlabel("Rentention Time (min)")
plt.ylabel("Intensity (10^6)")
plt.legend(["205nm", "250nm", "290nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone")
```

```python
x = np.linspace(0, 38, 9476)
plt.plot(x, resy[0:9476], color="blue") #205
plt.xlabel("Rentention Time (min)")
plt.ylabel("Intensity")
plt.legend(["205nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone at 205nm")
```

```python
plt.plot(resx[9476:(9476*2)], resy[9476:(9476*2)], color="green") #250
plt.xlabel("Rentention Time (min)")
plt.ylabel("Intensity")
plt.legend(["250nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone at 250nm")
```

```python
plt.plot(resx[9476*2:(9476)*3], resy[9476*2:(9476*3)], color="orange") #290
plt.xlabel("Rentention Time (min)")
plt.ylabel("Intensity")
plt.legend(["290nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone at 290nm")
```

```python
with open("overnight_oxy.txt") as f:
    lines = f.readlines()
    oresx=[]
    oresy=[]
    ocountres=[]
    app=False
    count=0
    for line in lines:
        if line.startswith("[PDA"):
            app=False
            ocountres.append(count)
            count=0
        if line.strip()=="":
            app=False
        if app:
            tmp=line.split("	")
            oresx.append(float(tmp[0]))
            oresy.append(float(tmp[1]))
            count +=1
        if line.startswith("R.Time"):
            app=True
    ocountres.append(count)
```

```python
plt.plot(oresx[(9476):(9476)*2], oresy[(9476):9476*2], color="blue") #205 overnight
plt.plot(oresx[0:(9476)], oresy[0:(9476)], color="green") #250 overnight
plt.plot(oresx[(9476)*2:(9476)*3], oresy[(9476)*2:(9476)*3], color="orange") #290 overnight
plt.xlabel("Rentention Time (min)")
plt.ylabel("Intensity (10^6)")
plt.legend(["205nm","250nm","290nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone(Overnight) at 250nm, 205nm, 290nm")
```
