```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open("0_5oxy_1percentDMSOWATER.txt") as f:
    lines = f.readlines()
    resx=[]
    resy=[]
    countres=[]
    app=False
    count=0
    for line in lines:
        if line.startswith("[PDA"):
            app=False
            countres.append(count)
            count=0
        if line.strip()=="":
            app=False
            continue
        if app:
            tmp=line.split("	")
            resx.append(float(tmp[0]))
            resy.append(float(tmp[1]))
            count +=1
        if line.startswith("R.Time"):
            app=True
```

```python
plt.plot(resx[0:9476], resy[0:9476], color="blue") #205
plt.plot(resx[9476:(9476*2)], resy[9476:(9476*2)], color="green") #250 
plt.plot(resx[9476*2:(9476)*3], resy[9476*2:(9476*3)], color="orange") #290
plt.xlabel("Retention Time (min)")
plt.ylabel("Intensity (10^6)")
plt.legend(["205nm", "250nm", "290nm"])
plt.title("HPLC Reading Analysis of 0.5 mg/ml Oxycodone")
```
